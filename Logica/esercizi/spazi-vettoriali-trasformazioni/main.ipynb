```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_endomorphism():
    """
    """
    # Definizione della trasformazione lineare (endomorfismo)
    transformation_matrix = np.array([[2, 0], [0, 1]])  # Esempio di matrice di trasformazione

    # Creazione di una griglia di vettori nello spazio vettoriale 2D
    x = np.linspace(-2, 2, 10)
    y = np.linspace(-2, 2, 10)
    X, Y = np.meshgrid(x, y)

    # Vettori originali
    vectors = np.stack([X.flatten(), Y.flatten()], axis=0)

    # Applicazione dell'endomorfismo
    transformed_vectors = transformation_matrix @ vectors

    # Plot dell'endomorfismo
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    # Plot vettori originali e trasformati
    axs[0].quiver(vectors[0], vectors[1], vectors[0], vectors[1], color='blue', angles='xy',
                  scale_units='xy', scale=1, alpha=0.5, label='Original vectors')
    axs[0].quiver(vectors[0], vectors[1], transformed_vectors[0], transformed_vectors[1],
                  color='red', angles='xy', scale_units='xy', scale=1, alpha=0.8, label='Transformed vectors')

    axs[0].axhline(0, color='black', linewidth=0.5)
    axs[0].axvline(0, color='black', linewidth=0.5)
    axs[0].set_xlim(-3, 3)
    axs[0].set_ylim(-3, 3)
    axs[0].set_aspect('equal', adjustable='box')
    axs[0].grid(True, linestyle='--', alpha=0.6)
    axs[0].set_title("Endomorphism of a 2D Vector Space")
    axs[0].set_xlabel('x')
    axs[0].set_ylabel('y')
    axs[0].legend()

    # Regole della definizione 2.4
    definition = """
    Definizione 2.4: Endomorfismo
    Sia V uno spazio vettoriale su un campo K. Un endomorfismo  un'applicazione
    lineare f: V -> V tale che:
    1. f(av + bw) = af(v) + bf(w), per ogni a, b  K e v, w  V.
    2. f  chiusa rispetto alle operazioni nello spazio V.
    """
    axs[1].text(0.5, 0.5, definition, fontsize=12, wrap=True, ha='center', va='center')
    axs[1].axis('off')
    axs[1].set_title("Definizione di Endomorfismo")

    plt.tight_layout()
    plt.show()

# Esegui la funzione per visualizzare l'endomorfismo e la definizione
visualize_endomorphism()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Funzione per rappresentare graficamente il Teorema 2.6
def visualize_isomorphism():
    """
    Visualizza l'isomorfismo tra End(V) e Mat_{n x n}(K).
    """
    # Spazio vettoriale V e una base
    basis = [np.array([1, 0]), np.array([0, 1])]  # Base canonica di R^2

    # Endomorfismo su V definito da una matrice
    transformation_matrix = np.array([[2, 1], [0, 1]])

    # Applica la trasformazione ai vettori della base
    transformed_basis = [transformation_matrix @ v for v in basis]

    # Visualizzazione dello spazio vettoriale e trasformazioni
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    # Grafico dello spazio vettoriale con la trasformazione
    for v, tv in zip(basis, transformed_basis):
        axs[0].quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original basis')
        axs[0].quiver(0, 0, tv[0], tv[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed basis')

    axs[0].axhline(0, color='black', linewidth=0.5)
    axs[0].axvline(0, color='black', linewidth=0.5)
    axs[0].set_xlim(-3, 3)
    axs[0].set_ylim(-3, 3)
    axs[0].set_aspect('equal', adjustable='box')
    axs[0].grid(True, linestyle='--', alpha=0.6)
    axs[0].set_title("Endomorfismo nello Spazio V")

    # Mostra la matrice associata
    axs[1].imshow(transformation_matrix, cmap='viridis', extent=(0, 2, 0, 2))
    for i in range(transformation_matrix.shape[0]):
        for j in range(transformation_matrix.shape[1]):
            axs[1].text(j + 0.5, i + 0.5, f"{transformation_matrix[i, j]}",
                        ha='center', va='center', color='white', fontsize=14)
    axs[1].set_title("Matrice Associata")
    axs[1].set_xticks([0.5, 1.5])
    axs[1].set_yticks([0.5, 1.5])
    axs[1].set_xticklabels(["x", "y"])
    axs[1].set_yticklabels(["x'", "y'"])
    axs[1].grid(False)

    plt.tight_layout()
    plt.show()

# Esegui la funzione
visualize_isomorphism()
```
